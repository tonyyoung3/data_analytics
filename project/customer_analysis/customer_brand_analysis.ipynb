```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.1f}%\n({v:d})'.format(p=pct, v=val)
    return my_autopct
```

```python
#Pack Size Preferences
pack_size_range = sales_data['PACK_SIZE'].max() - sales_data['PACK_SIZE'].min()
#pack_size_range

# Define the bin edges for the 10 groups for pack sizwe
num_groups = 10
bin_edges = [sales_data['PACK_SIZE'].min() + i * pack_size_range / num_groups for i in range(num_groups)] + [sales_data['PACK_SIZE'].max()]

# Define the bin labels for the 10 groups
bin_labels = [f'{bin_edges[i]}-{bin_edges[i+1]}' for i in range(len(bin_edges)-1)]

# Create a new column 'pack_size_group' based on the binning
sales_data['pack_size_group'] = pd.cut(sales_data['PACK_SIZE'], bins=bin_edges, labels=bin_labels)
#sales_data.head(5)


# Analyzing total sales by pack size
pack_size_preferences = sales_data.groupby('pack_size_group').agg(
    total_sales=pd.NamedAgg(column='TOT_SALES', aggfunc='sum')
).sort_values(by='total_sales', ascending=False)
pack_size_preferences
```

```python
user_group = customer_sales[['LYLTY_CARD_NBR', 'sales_group']]

df_with_group = sales_data.merge(user_group, on = 'LYLTY_CARD_NBR')
user_seg = df_with_group[['LYLTY_CARD_NBR','LIFESTAGE','PREMIUM_CUSTOMER','sales_group']]
user_seg = user_seg.drop_duplicates()
```

```python
# Group by 'premium' and count 'sales_group' occurrences
segment_analysis = user_seg.groupby('PREMIUM_CUSTOMER')['sales_group'].value_counts().unstack(fill_value=0)

# Calculate the percentage
group_percentage = segment_analysis.div(segment_analysis.sum(axis=1), axis=0) * 100
print(group_percentage)
```

```python
import numpy as np
import scipy.stats as stats

# Calculate the totals for each group and each premium group
row_totals = group_percentage.sum(axis=1)
column_totals = group_percentage.sum(axis=0)
grand_total = row_totals.sum()

# Calculate the expected counts for each cell under the null hypothesis
expected = np.outer(row_totals, column_totals) / grand_total

# Compute the chi-squared test statistic and p-value
chi2, p_value, dof, expected = stats.chi2_contingency(group_percentage)

print(f"Chi-squared Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
if p_value > 0.05:
    print(f"p value is {p_value} > 0.05, null hypothesis can not be rejected")
else:
    print(f"p value is {p_value} < 0.05, null hypothesis is rejected")
#print(expected)
#Even though the p test reject null hypothesis, but the difference is small.
```

```python
# example of cross brand selling
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = 'iframe'

customers_who_bought_brand_A = sales_data[sales_data['BRAND'] == 'KETTLE']['LYLTY_CARD_NBR'].unique()
other_brands_purchased = sales_data[sales_data['LYLTY_CARD_NBR'].isin(customers_who_bought_brand_A)]

brand_counts_excluding_A = other_brands_purchased[other_brands_purchased['BRAND'] != 'KETTLE']['BRAND'].value_counts()
brand_counts_excluding_A = brand_counts_excluding_A/20611

brand_counts_excluding_A_df = pd.DataFrame(brand_counts_excluding_A)
print('Perference of people who has bought Keetle')
print(brand_counts_excluding_A_df)
```

```python
labels = ['Kettle'] + brand_counts_excluding_A_df.index.tolist()  # ['A', 'B', 'C', 'D']
source = [0] * brand_counts_excluding_A_df.shape[0]  # [0, 0, 0] for 'A' as the source
target = list(range(1, brand_counts_excluding_A_df.shape[0] + 1))  # [1, 2, 3] for 'B', 'C', 'D'
values = brand_counts_excluding_A_df['count'].tolist()  # The flow values
```

```python
import matplotlib.cm as cm
# Choose a colormap (e.g., 'viridis', 'plasma', 'cividis')
colormap = cm.get_cmap('viridis')

# Normalize values for color mapping
norm = plt.Normalize(vmin=min(values), vmax=max(values))
colors = [colormap(norm(value)) for value in values]

# Convert RGBA colors to hex format for Plotly
colors = [f'rgba({int(r*255)}, {int(g*255)}, {int(b*255)}, {a})' for r, g, b, a in colors]
```

```python
#sankey chart
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels, 
    ),
    link=dict(
        source=source,
        target=target,
        value=values,
        color=colors,
    ))])

fig.update_layout(title_text=" Potential cross-brand appeal (Kettle)", font_size=10)
fig.update_layout(width=1000, height=600) 
fig.show()
```
